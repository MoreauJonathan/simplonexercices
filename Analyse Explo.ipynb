```python
import pandas as pd
#import de données depuis un fichier Excel dans une dataframe
nom_colonnes = ['consommation','cylyndres','deplacement','puissance',
                'poids','acceleration','annee','origine','nom']
data = pd.read_excel('//home//simplon//Téléchargements//autompg.xlsx',
                     index_col=None,header=None,names=nom_colonnes,skiprows=[0])
#Afficher les premieres ligne du Dataframe
print(data.head(),"\n")


#Afficher les dimensions du dataframe
print(data.shape)
print(data.shape[0])
print(data.shape[1],"\n")

#Information sur les données
print(data.info(),"\n")

#Description des données
print(data.describe(),"\n")

#Fréquences corisés
print(pd.crosstab(data['cylyndres'],data['annee']),"\n")

#Pourcentages corisés
print(pd.crosstab(data['cylyndres'],data['annee']),"\n")

#accés a une ou plusieurs colonnes 
print(data[['nom','consommation']].head,"\n")

#Affiche les valeurs 1O à 12
print(data['nom'][10:13],"\n")

#Accés à la valeur de la première ligne, première colonne
print(data.iloc[0:0],"\n")

#Accés à la valeur de la dernière ligne, première colonne
print(data.iloc[-1:0],"\n")

#Toutes les colonnes avec les lignes de 0 à 4
print(data.iloc[0:5,:],"\n")

#Liste des voitures dont l'année est de 80
print(data.loc[data['annee']==80,:].head,"\n")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

#import de données depuis un fichier Excel dans une dataframe
nom_colonnes = ['consommation','cylyndres','deplacement','puissance',
                'poids','acceleration','annee','origine','nom']
data = pd.read_excel('//home//simplon//Téléchargements//autompg.xlsx',
                     index_col=None,header=None,names=nom_colonnes,skiprows=[0])
#Pour que les graphiques apparaissent dans le notebook
%matplotlib inline

#histogramme de la puissance
data.hist(column = 'puissance')

#Nuage de points : consommation en fct de la puissance avec niveau de gris selon l'année
data.plot.scatter(x='puissance',y='consommation',c='annee')

pd.plotting.scatter_matrix(data.select_dtypes(exclude=['object']),figsize =(10,10))
```
